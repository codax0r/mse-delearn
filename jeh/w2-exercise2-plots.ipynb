```python
import numpy as np
from numpy import exp
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_der(z):
    sigmoid_z = sigmoid(z)
    return sigmoid_z * (1 - sigmoid_z)
```

```python
x = np.arange(-10, 10, .01)
```

## Exercise 2 c)

```python
fct = lambda z: -np.log(sigmoid(-z))

def fct_first_der(z):
    ez = np.exp(z)
    return ez / (1 + ez)

def fct_sec_der(z):
    ez = np.exp(z)
    return (ez / (1+ez)) - ((ez**2) / (1+ez)**2)

def plot(title, xlabel, ylabel, xs, ys, label=None):
    fig, ax = plt.subplots(figsize=(8, 6))
    _ = ax.plot(xs, ys, label=label)
    _ = ax.set_title(title)
    _ = ax.set_ylabel(ylabel)
    _ = ax.set_xlabel(xlabel)
    
    return fig, ax
```

```python
y_fct = fct(x)
y_fct_first_der = fct_first_der(x)
y_fct_sec_der = fct_sec_der(x)
```

```python
fig, ax = plot('fct(z)', 'z', 'fct(z)', x, y_fct, label='$\zeta(z)$')
_ = ax.plot(x, y_fct_first_der, label='$\zeta\'(z)$')
_ = ax.plot(x, y_fct_sec_der, label='$\zeta\'\'(z)$')
_ = ax.legend()
```

## Exercise 2f)

```python
def c1(x):
    return (sigmoid(x) - 1)**2
def c1_first_der(x):
    s = sigmoid(x)
    return 2 * s * (1-s) * (s-1)
def c1_sec_der(x):
    s = sigmoid(x)
    return (2 * s * (1-s) * (1-s) * (s-1) +
        (-2) * s * s * (1-s) * (s-1) +
        2 * s * (1-s) * s * (1-s))
```

```python
y_c1 = c1(x)
y_c1_first_der = c1_first_der(x)
y_c1_sec_der = c1_sec_der(x)
```

```python
fig, ax = plot('$c_1(x)$', 'x', '$c_1(x)$', x, y_c1, label='$c_1(x)$')
_ = ax.plot(x, y_c1_first_der, label='$c_1\'(x)$')
_ = ax.plot(x, y_c1_sec_der, label='$c_1\'\'(x)$')
_ = ax.legend()
```

#### Comment 
Since the second derivative is not alway positive, the function $c_1$ is not convex.

## Exercise 2g)

```python
def c2(x, y=1, w=1):
    return -y*np.log(sigmoid(w*x)) + (1-y)*np.log(1-sigmoid(w*x))
```

```python
y_c2 = c2(x)
```

```python
_ = plot('$c_2(x)$', 'x', '$c_2(x)$', x, y_c2)
```

![Plot](img/c2_plots.png "Plot of c2")

The plot here shows $c_2(x)$ (blue), $c_2'(x)$ (orange) and $c_2''(x)$ (yellow). We can see that the yellow line approximates zero but never touches it. So we can say that $c_2''(x) > 0$ and $c_2(x)$ is convex.
